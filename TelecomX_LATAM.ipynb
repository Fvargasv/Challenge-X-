### Total gastado

````python
total_gastado = pd.DataFrame(datos['Charges_Total'].groupby(datos['Churn']).sum().reset_index())
total_gastado
````

````python
# Definir colores para cada valor de churn
color_map = {
    0: "#1f77b4",  # azul
    1: "#ff7f0e"   # naranja
}

fig = px.bar(
    total_gastado,
    x="Churn",
    y="Charges_Total",
    text_auto=True,
    title="Total gastado por tipo de Churn",
    color="Churn",  # Se usa solo para aplicar color
    color_discrete_map=color_map,
    labels={  # Cambiar nombres de ejes
        "Churn": "Tipo de Churn",
        "Charges_Total": "DÃ³lares"
    }
             )

# Quitar la leyenda para que no aparezca como filtro
fig.update_layout(showlegend=False)
fig.update_coloraxes(showscale=False)

# Mostrar grÃ¡fico
fig.show()
````

## IntroducciÃ³n

El proyecto "Churn de Clientes" para Telecom X tiene como objetivo analizar la alta tasa de cancelaciones (evasiÃ³n) que enfrenta la empresa. A travÃ©s del anÃ¡lisis de datos, se busca identificar los factores que influyen en la pÃ©rdida de clientes para que el equipo de Data Science pueda desarrollar modelos predictivos y estrategias para reducir esta tasa.

En este proyecto, se recopilan, procesan y analizan datos relacionados con el comportamiento y caracterÃ­sticas de los clientes utilizando Python y sus principales bibliotecas.

## Limpieza y Tratamiento de Datos

### Base de datos

La Base de datos se entrega en un archivo con formato json y se solicita cargar datos directamente desde la API proporcionada, para esto se ejecuta el siguiente cÃ³digo:



```
import requests #para importar desde una API
import pandas as pd


url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
response = requests.get(url)
data = response.json()

datos = pd.DataFrame(data)

datos.head()
```



Es importante mencionar que al leer la base de datos json , vienen 6 columnas con la informaciÃ³n. El problema, es que 4 de estas columnas tienen diccionarios anidados, los cuÃ¡les son necesarios aperturar y esto se realiza normalizando el archivo json, con el siguiente cÃ³digo:



```
from pandas import json_normalize

datos = json_normalize(data, sep='_')
```

De esta forma, se obtiene un dataframe de 7267 filas y 21 columnas, con variables numÃ©ricas y categÃ³ricas.
