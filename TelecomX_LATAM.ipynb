````python
import plotly.express as px

# Crear columna con colores segÃºn Churn
total_gastado["Color"] = total_gastado["Churn"].map({
    0: "#87CEEB",  # skyblue
    1: "#F08080"   # lightcoral
})

# Graficar usando la columna de colores
fig = px.bar(
    total_gastado,
    x="Churn",
    y="Charges_Total",
    text_auto=True,
    title="Total gastado por tipo de Churn",
    color="Color",  # AquÃ­ se usa directamente el color
    labels={
        "Churn": "Tipo de Churn",
        "Charges_Total": "$"
    },
    color_discrete_map="identity"  # Usar los colores tal cual estÃ¡n en la columna
)

# Quitar leyenda y escala
fig.update_layout(showlegend=False)
fig.update_coloraxes(showscale=False)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.update_layout(
    plot_bgcolor="white",   # Fondo del Ã¡rea del grÃ¡fico
    paper_bgcolor="white"   # Fondo total de la figura
)

fig.show()
````

A travÃ©s del anÃ¡lisis de datos, se busca identificar los factores que influyen en la pÃ©rdida de clientes para que el equipo de Data Science pueda desarrollar modelos predictivos y estrategias para reducir esta tasa.

### Base de datos

La Base de datos se entrega en un archivo con formato json y se solicita cargar datos directamente desde la API proporcionada, para esto se ejecuta el siguiente cÃ³digo:



```

datos.head()

Es importante mencionar que al leer la base de datos json , vienen 6 columnas con la informaciÃ³n. El problema, es que 4 de estas columnas tienen diccionarios anidados, los cuÃ¡les son necesarios aperturar y esto se realiza normalizando el archivo json, con el siguiente cÃ³digo:



```

De esta forma, se obtiene un dataframe de 7267 filas y 21 columnas, con variables numÃ©ricas y categÃ³ricas.
