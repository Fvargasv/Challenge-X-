```python
import requests #para importar desde una API
import pandas as pd


url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
response = requests.get(url)
data = response.json()

datos = pd.DataFrame(data)

datos.head()
```

```python
datos.info()
```

##NormalizaciÃ³n archivo JSON

```python
from pandas import json_normalize

datos = json_normalize(data, sep='_')
datos.head(25)
```

```python
datos.info()
```

##DistribuciÃ³n de evasiÃ³n

```python
#cÃ³mo estÃ¡ distribuida la variable "churn" (evasiÃ³n) entre los clientes.
#Utiliza grÃ¡ficos para visualizar la proporciÃ³n de clientes que permanecieron y los que se dieron de baja.
import matplotlib.pyplot as plt

# Contar cuÃ¡ntos 1s y 0s hay en la columna Churn
conteo = datos['Churn'].value_counts().sort_index()
conteo
# Etiquetas para la torta
labels = ['No', 'Si']

# Colores opcionales
colors = ['skyblue', 'lightcoral']

# Crear grÃ¡fico de torta
plt.figure(figsize=(6,6))
plt.pie(conteo, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('ProporciÃ³n de clientes que SI dejaron la empresa vs\nlos que NO la dejaron  ')
plt.axis('equal')  # Para que se vea como un cÃ­rculo
plt.show()
```

```python
conteo
```

```python
import seaborn as sns

# Contar proporciÃ³n de 0 y 1
conteo = datos['Churn'].value_counts().sort_index()

# Estilo de Seaborn
sns.set_theme(style="whitegrid")

# Crear grÃ¡fico de torta con Matplotlib
plt.figure(figsize=(6,6))
plt.pie(
    conteo,
    labels=['0', '1'],
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Paired")[0:2]
)
plt.title('ProporciÃ³n de 1s y 0s en la columna')
plt.axis('equal')
plt.show()
```

### Total gastado

```python
total_gastado = pd.DataFrame(datos['Charges_Total'].groupby(datos['Churn']).sum().reset_index())
total_gastado
```

```python
fig = px.bar(total_gastado, text_auto = True ,title= 'Total gastado por tipo de Churn')
fig.show()
```

```python
# Definir colores para cada valor de churn
color_map = {
    0: "#1f77b4",  # azul
    1: "#ff7f0e"   # naranja
}

fig = px.bar(
    total_gastado,
    x="Churn",
    y="Charges_Total",
    text_auto=True,
    title="Total gastado por tipo de Churn",
    color="Churn",  # Se usa solo para aplicar color
    color_discrete_map=color_map,
    labels={  # Cambiar nombres de ejes
        "Churn": "Tipo de Churn",
        "Charges_Total": "DÃ³lares"
    }
             )

# Quitar la leyenda para que no aparezca como filtro
fig.update_layout(showlegend=False)

# Mostrar grÃ¡fico
fig.show()
```

### Tiempo de contrato

```python
tiempo_contrato = pd.DataFrame(datos['customer_tenure'].groupby(datos['Churn']).mean().round(1).reset_index())
tiempo_contrato
```

```python
fig = px.bar(
    tiempo_contrato,
    x="Churn",
    y="customer_tenure",
    text_auto=True,
    title="Meses promedio de duraciÃ³n del contrato por tipo de Churn",
    color="Churn",  # Se usa solo para aplicar color
    color_discrete_map=color_map,
    labels={  # Cambiar nombres de ejes
        "Churn": "Tipo de Churn",
        "customer_tenure": "Meses"
    }
)

# Quitar la leyenda para que no aparezca como filtro
fig.update_layout(showlegend=False)

# Mostrar grÃ¡fico
fig.show()
```
